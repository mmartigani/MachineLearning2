```python
df1=pd.read_csv('/content/Datos_Ventas_Tienda.csv').head(3)
df1
```

```python
df2=pd.read_csv('/content/Datos_Ventas_Tienda2.csv').head(2)
df2
```

```python
df3=pd.concat([df1,df2], ignore_index=True)
df3
```

```python
df3['Fecha']=pd.to_datetime(df3['Fecha'])
df3
```

```python
df3.info()
```

```python
producto = df3.groupby('Producto')['Cantidad'].sum()
producto = producto.sort_values(ascending=False)
producto.head(1)
```

```python
meses = []
for f in df3 ['Fecha']:
  meses.append(f.month)
df3["Meses"]=meses
df3
```

```python
ventas =df3.groupby('Meses')['Total Venta'].sum().sort_values(ascending=False)
ventas
```

```python
v_categoria = df3.groupby('Producto')['Total Venta'].sum()
v_categoria
```
