```python
n=np.random.randn(1000)
n
```

```python
plt.hist(n);
```

```python
plt.hist(n, bins=40);#bins nos permite ver la cantidad de barras
```

```python
plt.hist(n, bins=40, alpha=0.5); #alpha nos muestra la transparencia u opacidad
```

```python
plt.hist(n, bins=40, alpha=0.5, color='red');
```

```python
plt.hist(n, bins=40, alpha=0.5, color='red', edgecolor="green");#eggecolor, color de borde
```

```python
plt.hist(n, bins=40, alpha=0.5, color='red',edgecolor="blue", histtype='step');
#histtype borra la parte de adentro
```

```python
x1=np.random.randn(1000)
x2=np.random.randn(500)
x3=np.random.randn(100)
plt.hist(x1, histtype='stepfilled', alpha=0.3, bins=40)
plt.hist(x2, histtype='stepfilled', alpha=0.3, bins=40)
plt.hist(x3, histtype='stepfilled', alpha=0.3, bins=40);
```
