```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y,'o');
```

```python
aleatorio=np.random.RandomState(0)
marcadores = ['o','x','v','s','+','d']
```

```python
for marcador in marcadores:
  plt.plot(aleatorio.randn(5), aleatorio.rand(5), marcador)
```

```python
plt.plot(x,y,':ob')
```

```python
plt.plot(x,y,'-p')
```

```python
plt.plot(x,y,'-p', color="red", markersize=16)
```

```python
plt.plot(x,y,'-p', color="red", markersize=16, linewidth=4)
```

```python
plt.plot(x,y,'-p', color="red", markersize=16, linewidth=4, markerfacecolor="white")
```

```python
plt.scatter(x,y)
```
