```python
ruta=pd.read_csv('/content/Top-Películas.csv')
ruta.head()
```

```python
df_ordenados=ruta.sort_values(by='rating')
df_ordenados.head(5)
```

```python
df_ordenados=ruta.sort_values(by='rating', ascending=False)
df_ordenados.head(5)
```

```python
df_ordenados=ruta.sort_values(by=['rating', 'recaudación(M)'],ascending=False)
df_ordenados.head(5)
```

```python
df_agrupados = ruta.groupby('género')['rating'].mean()
df_agrupados
```

```python
df_agrupados=ruta.groupby('año')['recaudación(M)'].sum()
df_agrupados.sort_values(ascending=False).head(10)
```
