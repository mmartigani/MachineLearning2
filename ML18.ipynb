```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pinguinos=sns.load_dataset('penguins')
```

```python
sns.relplot(data=pinguinos,
            x='bill_length_mm',
            y='bill_depth_mm',
            hue='body_mass_g');
```

```python
sns.set_theme(style='dark')
sns.relplot(data=pinguinos,
            x='bill_length_mm',
            y='bill_depth_mm',
            hue='body_mass_g',
            palette='crest',
            marker='x',
            s=100);
```

```python
g=sns.set_theme(style='dark')
g=sns.relplot(data=pinguinos,
            x='bill_length_mm',
            y='bill_depth_mm',
            hue='body_mass_g',
            palette='crest',
            marker='x',
            s=100)
g.set_axis_labels('Largo del pico (mm)',
                  'Grosr del pico (mm)',
                  labelpad=10);
```

```python
g=sns.set_theme(style='dark')
g=sns.relplot(data=pinguinos,
            x='bill_length_mm',
            y='bill_depth_mm',
            hue='body_mass_g',
            palette='crest',
            marker='x',
            s=100)
g.set_axis_labels('Largo del pico (mm)',
                  'Grosr del pico (mm)',
                  labelpad=10)
g.legend.set_title("Masa\ncorporal en (g)");
```
