```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('/content/Iluminacion.csv')
df.head(2)
```

```python
fig, ax=plt.subplots()
hb=ax.hexbin(df['Latitud'], df['Longitud'],C=df['Luminosidad'], gridsize=50, cmap='Greens')
ax.set_facecolor('#000000')
plt.colorbar(hb,label="Nivel de sombras")
plt.xlabel("Latitud")
plt.ylabel('Longitud')
plt.title("Luminosidad");
```
