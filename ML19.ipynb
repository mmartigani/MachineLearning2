```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
p=sns.load_dataset('penguins')
sns.displot(data=p,
            x='flipper_length_mm',
            hue='species',
            multiple='stack'
            );
```

```python
p=sns.load_dataset('penguins')
sns.histplot(data=p,
            x='flipper_length_mm',
            hue='species',
            multiple='stack'
            );
```

```python
p=sns.load_dataset('penguins')
sns.displot(data=p,
            x='flipper_length_mm',
            hue='species',
            col='species');
```

```python
sns.relplot(data=p,
            x='flipper_length_mm',
            y='bill_length_mm',
            col='sex')
```

```python
graph=sns.relplot(data=p,
            x='flipper_length_mm',
            y='bill_length_mm',
            col='sex')
graph.set_axis_labels('Largo de aleta (mm)',
                      'Largo del pico (mm)');
```

```python
diamantes= sns.load_dataset('diamonds')
```

```python
diamantes.head()
```

```python
diamantes.describe()
```

```python
sns.set(style='whitegrid')
g=sns.relplot(data=diamantes,
              x='carat',
              y='price',
              kind='scatter',
              height=6,
              aspect=1.5)
g.fig.suptitle('relacion entre precio y peso de los diamntes',
               va='baseline',
               ha='center')
g.set_axis_labels("Peso en kilates", "Precio");
```

```python
g=sns.displot(data=diamantes,
              x='price',
              kind='hist',
              kde=True,
              color='blue',
              height=4,
              aspect=2)
g.fig.suptitle("Distribucion del precio del Diamante",
               va='baseline',
               ha='center')
g.set_axis_labels('Precio($)', 'Frecuencia');
```

```python
g=sns.histplot(data=diamantes,
              x='carat',
              kde=True,
              color='green',
              )
g.set_title("Distribucion del precio en quilates")
g.set(xlabel='Carat',ylabel='Frecuencia');
```

```python
g=sns.catplot(data=diamantes,
              x='cut',
              y='price',
              kind='box',
              height=5,
              aspect=2,
              order=['Fair', 'Good','Very Good', 'Premiun','Ideal'])
g.fig.suptitle("Comparacion entre el precio y la calidad de corte",
               va='baseline',
               ha='center')
g.set_axis_labels('Calidad del Corte', 'Precio($)');
```

```python
g=sns.jointplot(data=diamantes,
              x='depth',
              y='price',
              kind='kde',
              space=0,
              fill=True,
              color='purple')
g.fig.suptitle("Relacion entre la Profundidad  y el precio del Diamante",
               va='baseline',
               ha='center',
               fontsize=16)
g.set_axis_labels('profundidad del diamante', 'Precio($)');
```

```python
g=sns.pairplot(diamantes, corner=True)
g.fig.suptitle('Relaciones entre var numericas y el dataset',
               va='baseline',
               ha='center',
               fontsize=16);
```
