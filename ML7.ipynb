```python
import pandas as pd
```

```python
df1= pd.DataFrame({'ID':[1,2,3],
                   "Nombre":["ana","luis", "carlos"]})
df2=pd.DataFrame({'ID':[1,2,4],
                  "Edad":[25,30,22]})
```

```python
df1
```

```python
df2
```

```python
df_combinados= pd.merge(df1,df2, on='ID')
df_combinados
```

```python
df_indexados =pd.merge(df1,df2, left_index=True, right_index=True)
df_indexados
```

```python
df3=pd.DataFrame({'salario':[30000,45000,38000],
                  'antiguedad':[9,13,12]},
                 index=[1,2,3])
df3
```

```python
df4=pd.DataFrame({'ciudad':['madrid','barcelona', 'valencia'],
                 'jerarquia':['baja', 'alta','media']},
                 index=[1,2,4])
df4
```

```python
df_unido = df3.join(df4)
df_unido
```

```python
df5=pd.DataFrame({'nombre':['juan', 'gabriela','elena'],
                  'edad':[23,31,21]})
df5
```

```python
df6=pd.DataFrame({'nombre':['carmela', 'max','laura'],
                  'edad':[34,25,29]})
df6
```

```python
df_concatenados=pd.concat([df5,df6],ignore_index=True)
df_concatenados
```

```python
df_concatenados=pd.concat([df5,df6],keys=['df5', 'df6'])
df_concatenados
```

```python
df_concatenados=pd.concat([df5,df6],keys=['enero', 'diciembre'])
df_concatenados
```

```python
df_concatenados_2=pd.concat([df5,df6],axis=1)
df_concatenados_2
```
