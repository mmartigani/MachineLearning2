```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#REGRESION LOGISTICA = NOS DICE POR SI O NO CUAL ES LA MEJOR VAR
df=pd.read_csv('/content/datos_seguro.csv')
df.head()
# 0 es lo que no compraron
# 1 son los que compraron
```

```python
plt.scatter(df.edad, df.compra)
```

```python
X_train, X_test, y_train, y_test=train_test_split(df[['edad']], df['compra'],train_size=0.9)
#train_size=0.9 colocamos el 90% de los datos en entrenamiento
```

```python
X_train
```

```python
X_test
```

```python
modelo = LogisticRegression()
```

```python
modelo.fit(X_train,y_train)
```

```python
  modelo.score(X_train,y_train)
  #87% de aciertos
```

```python
datos_nuevos=pd.DataFrame({'edad':[25,35,45,55]})
```

```python
datos_nuevos
```

```python
probabilidades= modelo.predict_proba(datos_nuevos)
probabilidades
```

```python
prob_compra = probabilidades[:,1]
prob_compra
```

```python
plt.scatter(df.edad, df.compra)
plt.scatter(datos_nuevos['edad'],
            prob_compra,
            color='red');
```
