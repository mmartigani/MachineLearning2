```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.DataFrame({'etiquetas':['banana', 'manzana', 'pera', 'uva'],
                 'cantidad':[30,60,80,20]})
df
```

```python
plt.pie(df['cantidad']);
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas']);
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%');
#autopct='%1.1f%%' nos muestra el porcentaje en el grafico
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'],
        autopct='%1.1f%%',
        colors=['#87CEEB','#6495ED','#000080','#40D0E0']);
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'],
        autopct='%1.1f%%',
        colors=['#87CEEB','#6495ED','#000080','#40D0E0'],
        shadow=True);
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'],
        autopct='%1.1f%%',
        colors=['#87CEEB','#6495ED','#000080','#40D0E0'],
        shadow=True,
        explode=(0,0,0,0.5));
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'],
        autopct='%1.1f%%',
        colors=['#87CEEB','#6495ED','#000080','#40D0E0'],
        shadow=True,
        explode=(0,0,0,0.5))
plt.title("Frutas");
```
