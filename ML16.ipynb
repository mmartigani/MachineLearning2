```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
propinas=sns.load_dataset('tips')
propinas
```

```python
sns.relplot(data=propinas,
            x='total_bill',
            y='tip')
```

```python
sns.relplot(data=propinas,
            x='total_bill',
            y='tip',
            kind='line')
```

```python
sns.relplot(data=propinas,
            x='total_bill',
            y='tip',
            kind='line',
            hue='smoker')
```

```python
sns.relplot(data=propinas,
            x='total_bill',
            y='tip',
            kind='line',
            hue='smoker',
            style='smoker')
```

```python
sns.relplot(data=propinas,
            x='total_bill',
            y='tip',
            kind='line',
            hue='smoker',
            style='smoker',
            size='size')
```

```python
sns.relplot(data=propinas,
            x='total_bill',
            y='tip',
            hue='smoker',
            style='smoker',
            size='size')
```

```python
sns.relplot(data=propinas,
            x='total_bill',
            y='tip',
            hue='smoker',
            style='smoker',
            size='size',
            col='time')
```

```python
sns.relplot(data=propinas,
            x='total_bill',
            y='tip',
            legend=None,
            hue='smoker',
            style='smoker',
            size='size', col='time')
```
