```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import linear_model
```

```python
# APRENDIZAJE SUPERVISADO = TIENEN DATOS LOS APRENDAN INGRESAN NUEVOS DATOS
# Y APRENDEN DE LOS VISTOS Y PREDICE SEGUN LOS DATOS ANTES INGRESADOS APRENDIDOS
#REGRESION LINEAL, LOGISTICA, ARBOLES DECISION, BOSQUES ALEATORIOS
#-------------------------------------------
# APRENDIZAJE NO SUPERVISADO
# APRENDIZAJE POR REFORZAMIENTO
```

```python
#REGRESION LINEAL = PREDICCION ENTRE LAS VAR DE ENTRADA Y LAS DE SALIDA
df=pd.DataFrame()
df['Area']=[2600,3000,3200,3600,4000]
df['Precio']=[550000,565000,610000,680000,725000]
```

```python
plt.scatter(df['Area'], df['Precio']);
```

```python
X=df[['Area']] #var independiente ya conocemos
```

```python
y=df['Precio'] #esta var tenemos que predecir
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X,y)
```

```python
modelo.predict([[3300]])
```

```python
plt.plot(df['Area'], modelo.predict(X))
plt.scatter(df['Area'], df['Precio']);
```
