```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
propinas = sns.load_dataset('tips')
```

```python
sns.catplot(data = propinas,
            kind='strip',
            x='day',
            y='total_bill',
            hue='smoker');
```

```python
sns.catplot(data = propinas,
            kind='swarm',
            x='day',
            y='total_bill',
            hue='smoker');
```

```python
sns.catplot(data = propinas,
            kind='box',
            x='day',
            y='total_bill',
            hue='smoker');
```

```python
sns.catplot(data = propinas,
            kind='violin',
            x='day',
            y='total_bill',
            hue='smoker');
```

```python
sns.catplot(data = propinas,
            kind='point',
            x='day',
            y='total_bill',
            hue='smoker');
```

```python
sns.catplot(data = propinas,
            kind='bar',
            x='day',
            y='total_bill',
            hue='smoker');
```

```python
pinguinos=sns.load_dataset('penguins')
```

```python
sns.jointplot(data=pinguinos,
              x='flipper_length_mm',
              y='bill_length_mm',
              hue='species');
```

```python
sns.pairplot(data=pinguinos,
             hue='species');
```

```python
g=sns.PairGrid(pinguinos,
             hue='species');
```

```python
g=sns.PairGrid(pinguinos,
             hue='species',
             corner=True)
g.map_lower(sns.kdeplot,
            hue=None,
            level=5,
            color='.2');
```

```python
g=sns.PairGrid(pinguinos,
             hue='species',
             corner=True)
g.map_lower(sns.kdeplot,
            hue=None,
            level=5,
            color='.2')
g.map_lower(sns.scatterplot,
             marker='+');
```

```python
g=sns.PairGrid(pinguinos,
             hue='species',
             corner=True)
g.map_lower(sns.kdeplot,
            hue=None,
            level=5,
            color='.2')
g.map_lower(sns.scatterplot,
             marker='+')
g.map_diag(sns.histplot,
           element='step',
           linewidth=0,
           kde=True);
```

```python
g=sns.PairGrid(pinguinos,
             hue='species',
             corner=True)
g.map_lower(sns.kdeplot,
            hue=None,
            level=5,
            color='.2')
g.map_lower(sns.scatterplot,
             marker='+')
g.map_diag(sns.histplot,
           element='step',
           linewidth=0,
           kde=True)
g.add_legend(frameon=True);
```

```python
g=sns.PairGrid(pinguinos,
             hue='species',
             corner=True)
g.map_lower(sns.kdeplot,
            hue=None,
            level=5,
            color='.2')
g.map_lower(sns.scatterplot,
             marker='+')
g.map_diag(sns.histplot,
           element='step',
           linewidth=0,
           kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61,.6));
```
