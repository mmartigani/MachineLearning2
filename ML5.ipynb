```python
data1={"id_producto":[1001,1002,1003,1003],
       "cantidad_vendida":[30, None,25,25],
       "precio":[20.5,15.0,None,22.5]}
data1
```

```python
df1=pd.DataFrame(data1)
df1
```

```python
#vemos datos nulos
#los True son los datos nulos
df1.isnull()
```

```python
df1.isnull().sum()
```

```python
df_eliminados = df1.dropna()
df_eliminados
```

```python
valores_nuevos={"cantidad_vendida":0,"precio":df1["precio"].mean()}
valores_rellenados = df1.fillna(valores_nuevos)
valores_rellenados
```

```python
valores_rellenados["cantidad_vendida"]=valores_rellenados["cantidad_vendida"].astype(int)
valores_rellenados
```

```python
valores_unicos = valores_rellenados.drop_duplicates(subset="id_producto")
valores_unicos
```
